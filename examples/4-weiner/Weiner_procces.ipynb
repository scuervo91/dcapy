```python
import os
from dcapy.weiner import Weiner
from dcapy.dca import ProbVar
import numpy as np 
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
import seaborn as sns 
from scipy import stats
```

```python
x = Weiner()

df = x.brownian_motion(500,4)
#dfa = x.brownian_motion(5000,3,interval=0.6)
fig, ax = plt.subplots()
df.plot(ax=ax,legend=False)
#dfa.plot(ax=ax)
```

```python
x1 = Weiner(initial_condition=100,drift=0.5, freq='A')

fig,ax = plt.subplots(figsize=(12,7))
df1 = x1.brownian_motion(50*12,10, freq_output='M')
df1.plot(ax=ax,legend=False)
df1a = x1.brownian_motion(50*12,3, freq_output='M',interval=0.1)
df1a.plot(ax=ax)
```

```python
x2 = Weiner(
    initial_condition=0,
    generator = ProbVar(dist='norm',kw={'loc':0,'scale':0.2887}, seed=910821),
    drift=0.2/12,
    freq='M')

fig,ax = plt.subplots(figsize=(12,7))
df2 = x2.brownian_motion(50*12,10, freq_output='M')
df2.plot(ax=ax,legend=False)
df2a = x2.brownian_motion(50*12,3, freq_output='M',interval=0.1)
df2a.plot(ax=ax)
```

```python
x3 = Weiner(
    initial_condition=10,
    generator = ProbVar(dist='norm',kw={'loc':0,'scale':0.26}, seed=9113),
    drift=0.03,
    freq='A')

fig,ax = plt.subplots(figsize=(12,8))
df3 = x3.geometric_brownian_motion(50*12,20, freq_output='M')
df3.plot(ax=ax,legend=False)
df3a = x3.geometric_brownian_motion(50*12,3, freq_output='M',interval=0.1)
df3a.plot(ax=ax)
```

```python
#x4 = Weiner(initial_condition=10,generator=stats.norm, mu=0.03/12, kw_generator={'scale':np.sqrt(np.power(0.26,2)/12)},freq_mu='A', seed=9)

x4 = Weiner(
    initial_condition=10,
    generator = ProbVar(dist='norm',kw={'loc':0,'scale':np.sqrt(np.power(0.26,2)/12)}, seed=9),
    drift=0.03/12,
    freq='A')

fig,ax = plt.subplots(figsize=(12,8))
df4 = x4.geometric_brownian_motion(20*12,20, freq_output='A')
df4.plot(ax=ax,legend=False)
df4a = x4.geometric_brownian_motion(20*12,3, freq_output='A',interval=0.1)
df4a.plot(ax=ax, linewidth=4)
```
